```python
import math
import numpy as np
import matplotlib.pyplot as plt

def apply_x0(f,x0):
    return lambda x: f(x,x0)

def test_fun(x,x0):
    return 10*(1+(x-x0)**2/3-math.cos(3*math.pi*(x-x0)))

def plot(f,a,b,d): # f - funkcja, a - początek przedziału, b - koniec przedziału, d - średnica podziału
    x=np.linspace(a,b,int((b-a)//d))
    y=np.vectorize(f)(x)
    plt.plot(x,y)
    
def find_min_naive(f,a,b,d): # f - funkcja, a - początek przedziału, b - koniec przedziału, d - średnica podziału
    x=np.linspace(a,b,int((b-a)//d))
    y=np.vectorize(f)(x)
    return min(y)

plot(apply_x0(test_fun,2),-10,10,0.01)
print(find_min_naive(apply_x0(test_fun,2),-10,10,0.01))
```

```python
import random
import math

# ----------------------------------------------------------------------------
dokladnosc = 1000
XDprzedzal_max = 300
XDprzedzal_min = -300
#------------------------------------------------------------------------------
def wartosc_f(f,wartosc,start = -10,end = 10,step = 0.2):
    temp_start = start
    while(temp_start <= end):
        if(wartosc >= f(temp_start)): return True
        temp_start += step
    return False
    
def minfun(f,przedzial_min,przedzial_max,dokladnosc):
    dokladnosc-=1
    if(dokladnosc == 0): return (przedzial_min+przedzial_max)/2
    if(wartosc_f(f,przedzial_max) == False): return minfun(f,przedzial_max,przedzial_max + 100,dokladnosc) 
    if(wartosc_f(f,przedzial_min) == True): return minfun(f,przedzial_min - 100,przedzial_min,dokladnosc) 
    temp = random.uniform(przedzial_max,przedzial_min)
    if(wartosc_f(f,temp) == True): return minfun(f,przedzial_min,temp,dokladnosc)
    if(wartosc_f(f,temp) == False): return minfun(f,temp,przedzial_max,dokladnosc)


def W(x):
    return x*x-8*x+2

print(minfun(W,XDprzedzal_min,XDprzedzal_max,dokladnosc,))

def G(x):
    return 10*(1+(x)**2/3-math.cos(3*math.pi*(x)))

print(minfun(G,XDprzedzal_min,XDprzedzal_max,dokladnosc,))

def W3(x):
    return x*x*x-8*x+2

print(minfun(W3,XDprzedzal_min,XDprzedzal_max,dokladnosc,))

```
